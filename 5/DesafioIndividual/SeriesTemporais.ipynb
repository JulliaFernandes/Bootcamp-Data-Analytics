```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
vendas = pd.read_csv('vendas.csv', delimiter=',')
```

```python
vendas.head()
```

```python
## Criando um grfico de linhas de vendas ao longo do tempo
plt.figure(figsize=(12, 6))
plt.plot(vendas['date'], vendas['sales'])
plt.title('Vendas ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
## Criando um grfico de decomposio sazonal 
decomposition = seasonal_decompose(vendas['sales'], model='additive', period=12)
decomposition.seasonal.plot()
plt.title('Sazonalidade')
plt.xlabel('Data')
plt.ylabel('Sazonalidade')
plt.show()
```

```python
## Criando um grfico interarivo que permita a explorao dos dados de vendas ao kongo do tempo
import plotly.express as px

fig = px.line(vendas, x='date', y='sales', title='Vendas ao longo do tempo')
fig.show()
```
