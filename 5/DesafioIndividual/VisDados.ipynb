```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd
```

```python
saude = pd.read_csv('saude.csv')
```

```python
saude.head()
```

```python
## Criando um histograma com a variável age 
plt.hist(saude['age'], bins=20, edgecolor='black')
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
## Criando um gráfico de dispersão mostrando a relação entre 'height' e 'weight'

sns.scatterplot(x='height', y='weight', data=saude)
plt.title('Relação entre altura e peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.show()
```

```python
## Criando um gráfico interativo de barras mostrando a contagem de indivíduos por 'gender'
import plotly.express as px

fig = px.bar(saude, x='gender', color='gender', title="Contagem de Indivíduos por Gênero",
             labels={'gender': 'Gênero'})                           line=dict(width=2, color='DarkSlateGrey')))

fig.show()
```
