### Questão 9:

```python
### importa o pacote
# !pip install statsmodels
import statsmodels.api as sm

### define a variável explicativa e adiciona uma constante
X = sm.add_constant(df2['Fuel Tank Capacity'])


### define a variável de interesse

y = df2['Price']
```

```python
# Fit do modelo: aqui o python ira calcular os valores de intercepto e coeficiente angular estimados
model = sm.OLS(y, X).fit()
```

```python
## gera a tabela de regressão
print(model.summary())
```

O modelo estimado foi:

-3.177⁶ + 9.375⁴*(capacidade do tanque)


```python
## gerando o gráfico da reta estimada

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df2, x='Fuel Tank Capacity', y='Price', color='blue', alpha=0.5)

# a funcao regplot gera a regressao e a plot da reta
sns.regplot(data=df2, x='Fuel Tank Capacity', y='Price', scatter=False, color='red')

# definindo titulos
plt.title('Regressão do Preço do Veículo vs Capacidade do tanque')
plt.xlabel('Capacidade do tanque')
plt.ylabel('Preço')

plt.show()
```

### Questão 10:

```python
# salva o dado do resíduo
residuals = model.resid
# calcula ovalor predito
predicted_values = model.fittedvalues ## o .fittedvalues serve para pegar os valores preditos

plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, color='blue', alpha=0.5)

# adiciona linha
plt.axhline(y=0, color='red', linestyle='--')

# titulos
plt.title('Capacidade do tanque vs Preço')
plt.xlabel('Capacidade do tanque')
plt.ylabel('Preço')

plt.show()

## esse grafico parece estar disperso e nao parece ter um padrão/linearidade, o que é bom
```
